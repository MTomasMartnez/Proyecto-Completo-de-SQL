### INDEX

#### 驴Qu茅 es un ndice (Index) en una Base de Datos?

Un **铆ndice** en una base de datos es como un 铆ndice de un libro. Cuando tienes un libro muy grande y necesitas encontrar un capitulo espec铆fico. En lugar de leer p谩gina por p谩gina, vas al 铆ndice del libro, y buscas el capitulo y r谩pidamente encuentras la p谩gina exacta donde se encuentra. En las bases de datos, un 铆ndice funciona de manera similar.

**Funci贸n de los ndices:**
- Los 铆ndices ayudan a **acelerar** la b煤squeda de datos dentro de una tabla.
- Facilitan las consultas al permitir que el sistema encuentre datos sin tener que revisar cada fila de la tabla (esto se llama un "scan" completo de la tabla).
- Se crean en una o m谩s columnas de una tabla, dependiendo de qu茅 columnas se usan frecuentemente en las consultas.

**Ventajas de los ndices:**
- **Mejoran el rendimiento** de las consultas SELECT.
- Son 煤tiles cuando necesitas ordenar datos o buscar filas espec铆ficas r谩pidamente.

**Desventajas:**
- Los 铆ndices ocupan **espacio** adicional en disco.
- Las operaciones de escritura (INSERT, UPDATE, DELETE) pueden ser m谩s lentas porque los 铆ndices deben actualizarse.
